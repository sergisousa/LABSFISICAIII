```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
## RECOLHER OS DADOS PARA O PYTHON

dados = np.zeros([10,3500])
tt, yy = [], []
with open('FLUORESCENCIA0.dat') as file:
        for line in file:
            # print(line[0])
            if line[0]=='0':
                yy.append(float(line.split('\t')[1]))
                tt.append(float(line.split('\t')[0].split(':')[2]))
yy = np.array(yy)
dados[0] = yy

for i in range(1,10):
    yy = []
    with open(f'FLUORESCENCIA{i}.dat') as file:
        for line in file:
            if line[0]=='0':
                yy.append(float(line.split('\t')[1]))
    yy = np.array(yy)
    dados[i] = yy
            

t0 = np.array(tt)
y00 = np.array(dados[0])
y01 = np.array(dados[1])
y02 = np.array(dados[2])
y03 = np.array(dados[3])
y04 = np.array(dados[4])
y05 = np.array(dados[5])
y06 = np.array(dados[6])
y07 = np.array(dados[7])
y08 = np.array(dados[8])
y09 = np.array(dados[9])
```

```python
## ENSAIO 0

def exponential_function(x, A, B):
    return A * np.exp(-x*B)

# Dados a usar na regresso exponencial
t = t0[1300:2000]-t0[1300]
y = y00[1300:2000] - np.min(y00)

params, covariance = curve_fit(exponential_function, t, y)

A0, B0 = params

## TAU

print(1/B0)

## GRFICO

plt.figure(figsize=(16,9))
plt.plot(t0-t0[1300], y00-np.min(y00))
aj0 = A0*np.exp(-B0*t0)
plt.plot(t0,aj0)
```

```python
## COMPARAO

plt.figure(figsize=(16,9))
plt.plot(t0, aj0, label='Ensaio 0') # Vemos que  outlier, mas  por causa do A0. Exclumos.
plt.plot(t0, aj1, label='Ensaio 1')
plt.plot(t0, aj2, label='Ensaio 2')
plt.plot(t0, aj3, label='Ensaio 3')
plt.plot(t0, aj4, label='Ensaio 4')
plt.plot(t0, aj5, label='Ensaio 5')
plt.plot(t0, aj6, label='Ensaio 6')
plt.plot(t0, aj7, label='Ensaio 7')
plt.plot(t0, aj8, label='Ensaio 8')
plt.plot(t0, aj9, label='Ensaio 9')
plt.title("Ajustes Exponenciais para diferentes ensaios", fontsize=20)
plt.xlabel("Tempo (s)",fontsize=16)
plt.ylabel("Sinal (Volts)",fontsize=16)
plt.legend(fontsize=12)
plt.show()

## CALCULAR Tau COM INCERTEZA (SEM ENSAIO 0)
Bs = np.array([B1,B2,B3,B4,B5,B6,B7,B8,B9])
taus = 1/Bs
tau = np.average(taus)
u_tau = np.std(taus)/np.sqrt(taus.shape[0])
print(f' = ({tau}  {u_tau}) s')
print(f' = {tau} s  {(u_tau/tau)*100} %')
print()
print()

## FUNO DE DECAIMENTO (SEM ENSAIO 0)
As = np.array([A1,A2,A3,A4,A5,A6,A7,A8,A9]) # Ao retirar o A0 a incerteza vai de 4 % para 2 %
Bs = np.array([B1,B2,B3,B4,B5,B6,B7,B8,B9])
A = np.average(As)
B = np.average(Bs)
u_A = np.std(As)/np.sqrt(As.shape[0])
u_B = np.std(Bs)/np.sqrt(Bs.shape[0])
print(f'Obtivemos ento a funo de decaimento:')
print(f'I(t) = {A} e^(-{B} t)')
print()
print('Em que:')
print(f'I0 = ({A}  {u_A}) s = {A} s  {(u_A/A)*100} %')
print(f'1/tau = ({B}  {u_B}) s = {B} s  {(u_B/B)*100} %')

plt.figure(figsize=(16,9))
plt.plot(t0 - t0[1350], dados[0] - np.min(y00), label="Ensaio 0")
plt.plot(t0 - t0[1350], dados[1] - np.min(y00), label="Ensaio 1")
plt.plot(t0 - t0[1350], dados[2] - np.min(y00), label="Ensaio 2")
plt.plot(t0 - t0[1350], dados[3] - np.min(y00), label="Ensaio 3")
plt.plot(t0 - t0[1350], dados[4] - np.min(y00), label="Ensaio 4")
plt.plot(t0 - t0[1350], dados[5] - np.min(y00), label="Ensaio 5")
plt.plot(t0 - t0[1350], dados[6] - np.min(y00), label="Ensaio 6")
plt.plot(t0 - t0[1350], dados[7] - np.min(y00), label="Ensaio 7")
plt.plot(t0 - t0[1350], dados[8] - np.min(y00), label="Ensaio 8")
plt.plot(t0 - t0[1350], dados[9] - np.min(y00), label="Ensaio 9")
plt.plot(t0, A*np.exp(-B*t0), color='black', label=r'$I(t) = 0.04220 e^{-143.47 t}$')
plt.title("Sobreposio dos Ensaios", fontsize=20)
plt.xlabel("Tempo (s)",fontsize=16)
plt.ylabel("Sinal (Volts)",fontsize=16)
plt.legend(fontsize=12)
plt.show()
```
