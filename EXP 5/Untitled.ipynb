```python
from scipy.stats import poisson
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Ambiente

```python
lista=[]
comb = []
skips = 0
xvals = []
yvals = []
yvals = []

with open("AMB1.TXT", "r") as data:
    for line in data:
        if skips<23:
            skips+=1
            continue
        linha = line.split("\t")
        if len(linha) > 1:
            x = float(linha[0])       
            y = float(linha[1])
            comb.append((x,y))
            xvals.append(x)
            yvals.append(y)

tam=len(yvals) 
med = sum(yvals)/len(yvals)
medl= np.copy(xvals)
medl.fill(med)
plt.figure(figsize=(12,8))
plt.title(f"N(T) - n={tam}", fontsize=20)
plt.ylabel("N (s.u)", fontsize=16)
plt.xlabel("T (segundos)", fontsize=16)
plt.scatter(xvals,yvals, label="Valores Experimentais")
plt.plot(xvals,medl, label=f"Mdia: {med}", color="#ff0000", linestyle="--")
plt.legend()
plt.show()
plt.figure(figsize=(12,8))
plt.hist(yvals,int(max(yvals)-min(yvals)), label="Valores Experimentais")
xvals = np.arange(min(yvals),max(yvals))
yvals = poisson.pmf(xvals,med) * len(yvals)
medl = [med,med]
ymedl = [0,max(yvals)]
plt.title(f"Frequncia(N) - n ={tam}", fontsize=20)
plt.ylabel("Frequncia (s.u)", fontsize=16)
plt.xlabel("N (s.u)", fontsize=16)
plt.plot(xvals,yvals, label="Distribuio de Poisson")
plt.plot(medl,ymedl, label=f"Mdia: {med}", color="#ff0000",linestyle="--")
plt.legend()
plt.show()
```
