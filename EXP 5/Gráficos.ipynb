```python
from scipy.stats import poisson
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from scipy.optimize import curve_fit
```

## Ambiente

```python
lista=[]
comb = []
skips = 0
xvals = []
yvals = []
yvals = []

with open("AMB1.TXT", "r") as data:
    for line in data:
        if skips<23:
            skips+=1
            continue
        linha = line.split("\t")
        if len(linha) > 1:
            x = float(linha[0])       
            y = float(linha[1])
            comb.append((x,y))
            xvals.append(x)
            yvals.append(y)

tam=len(yvals) 
med = sum(yvals)/len(yvals)
MEDAMB = med
medl= np.copy(xvals)
medl.fill(med)

incmed = np.std(yvals)/np.sqrt(len(yvals))
pincmed = incmed/med*100
print(f"MÃ©dia:{med}\nIncerteza: {incmed:.02} Incerteza (%): {pincmed:.03}")

plt.figure(figsize=(12,8))
plt.title(f"Atividade Ambiente - N(T) - n={tam}", fontsize=20)
plt.ylabel("N (s.u)", fontsize=16)
plt.xlabel("T (segundos)", fontsize=16)
plt.scatter(xvals,yvals, label="Valores Experimentais")
plt.plot(xvals,medl, label=f"MÃ©dia: {med}", color="#ff0000", linestyle="--")
plt.legend(fontsize=14)
plt.show()
plt.figure(figsize=(12,8))
plt.hist(yvals,int(max(yvals)-min(yvals)), label="Valores Experimentais", rwidth=0.85)
xvals = np.arange(min(yvals),max(yvals)+1)
yvals = poisson.pmf(xvals,med) * len(yvals)
medl = [med,med]
ymedl = [0,max(yvals)]
plt.title(f"Atividade Ambiente - FrequÃªncia(N) - n ={tam}", fontsize=20)
plt.ylabel("FrequÃªncia (s.u)", fontsize=16)
plt.xlabel("N (s.u)", fontsize=16)
plt.plot(xvals,yvals, label="DistribuiÃ§Ã£o de Poisson")
plt.plot(medl,ymedl, label=f"MÃ©dia: {med}", color="#ff0000",linestyle="--")
plt.legend(fontsize=14)
plt.show()
```
